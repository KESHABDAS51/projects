The objective of the dataset is to predict whether the patient has diabetes or not. There are several predictor(independent) variables like pregnencies, Blood pressure,skin thickness insulin,BMI,pedigree function and age. and there is one outcome which says diabetic or not

# checking the no. of zero values in the dataset

```python
print('Number of zero values in Glucose : ',df[df['Glucose']==0].shape[0])
print('Number of zero values in BloodPressure : ',df[df['BloodPressure']==0].shape[0])
print('Number of zero values in SkinThickness : ',df[df['SkinThickness']==0].shape[0])
print('Number of zero values in Insulin : ',df[df['Insulin']==0].shape[0])
print('Number of zero values in BMI : ',df[df['BMI']==0].shape[0])
```

```python
#replacing zero values with mean value
df['Glucose']=df['Glucose'].replace(0,df['Glucose'].mean())
df['BloodPressure']=df['BloodPressure'].replace(0,df['BloodPressure'].mean())
df['SkinThickness']=df['SkinThickness'].replace(0,df['SkinThickness'].mean())
df['Insulin']=df['Insulin'].replace(0,df['Insulin'].mean())
df['BMI']=df['BMI'].replace(0,df['BMI'].mean())
```

# splitting the data into independent(X) and target(Y) variable

```python
target='Outcome'
Y=df[target]
X=df.drop(target,axis=1)
```

```python
X.head(3)
```

```python
Y.head(3)
```

# Train Test split

```python
from sklearn.model_selection import train_test_split
X_train,X_test,Y_train,Y_test=train_test_split(X,Y,test_size=0.2,random_state=7)
```

# Feature Scaling¶

```python
#apply standard scale
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
scaler.fit_transform(X_train)
scaler.transform(X_test)
```

```python
X_train.shape,X_test.shape
```

# Build the Classification algorithm(Decision Tree)

```python
from sklearn.tree import DecisionTreeClassifier
dt=DecisionTreeClassifier()
dt.fit(X_train,Y_train)
```

# Making Prediction on Test Data

```python
dt_pred=dt.predict(X_test)
```

# Model Evaluation

```python
#train test score of Decision Tree
from sklearn.metrics import accuracy_score
print("train accuracy of Decision Tree: ",dt.score(X_train,Y_train)*100)
print("test accuracy of Decision Tree: ",dt.score(X_test,Y_test)*100)
print("accuracy of the Decision Tree Model: ",accuracy_score(Y_test,dt_pred)*100)
```
